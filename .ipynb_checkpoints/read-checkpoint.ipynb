```python
import pandas as pd
import numpy as np
import random
import string
np.random.seed(56)
```

```python
def id ():
    return ''.join(random.choice(string.ascii_letters) for i in range(5))
```

```python
def score():
    return np.random.randint(0, 100)
```

```python
df = pd.DataFrame({"국어":[score() for i in range(10)], "영어":[score() for i in range(10)], 
              "수학":[score() for i in range(10)], "사회":[score() for i in range(10)]}, 
             index=[id() for i in range(10)])
```

```python
df
```

```python
#과학이 추가되었습니다. 기존 학생들에게 각 랜덤으로 추가하시오.
```

```python
df.loc[:,'과학'] = pd.Series([score() for i in range(1, 11)], index=df.index)
```

```python
df
```

```python
# 각 학생들의 점수의 총합을 표현하는 컬럼을 추가하세요.
```

```python
df.loc['과목총점'] = [df['국어'].sum(),df['영어'].sum(),df['수학'].sum(),df['사회'].sum(),df['과학'].sum()]
```

```python
df
```

```python
df['총점'] = df.sum(axis=1)
```

```python
df
```

```python
df.sort_values(by='총점', ascending=False)
```

```python
df.drop("과목총점")
```
